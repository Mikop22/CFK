```python
import uuid
import psycopg2  # or your_db_connector

def get_target_menu_ids(connection, restaurant_id, subsection_id=None):
    """Get menu IDs based on restaurant_id and optional subsection_id"""
    cursor = connection.cursor()
    try:
        if subsection_id is not None:
            query = """
                SELECT menu_id
                FROM menus
                WHERE restaurant_id = %s
                AND subsection_id = %s
                AND DATE(created_at) = CURRENT_DATE
                ORDER BY created_at DESC;
            """
            cursor.execute(query, (restaurant_id, subsection_id))
        else:
            query = """
                SELECT menu_id
                FROM menus
                WHERE restaurant_id = %s
                AND DATE(created_at) = CURRENT_DATE
                ORDER BY created_at DESC;
            """
            cursor.execute(query, (restaurant_id,))

        results = cursor.fetchall()
        return [row[0] for row in results]
    except Exception as e:
        print(f"Error fetching target menu IDs: {e}")
        return []
    finally:
        cursor.close()

```

```python
TEMPLATE_MENU_ID = "ec08f5af-ddb4-445d-a74b-a6d2f1608e6c"  # ID of your fully configured menu

# Database connection parameters
DB_CONFIG = {
    'dbname': "your_db",
    'user': "your_user",
    'password': "your_password",
    'host': "your_host"
}

RESTAURANT_ID = "your-restaurant-id-here"
SUBSECTION_ID = "your-subsection-id-here"  # or None if you want all menus for the restaurant

# --- Script Logic ---
conn = None
cur = None

try:
    # Establish database connection
    conn = psycopg2.connect(**DB_CONFIG)
    cur = conn.cursor()

    # Get target menu IDs
    TARGET_MENU_IDS = get_target_menu_ids(conn, RESTAURANT_ID, SUBSECTION_ID)

    if not TARGET_MENU_IDS:
        print("No target menus found. Exiting.")
        exit()

    print(f"Found {len(TARGET_MENU_IDS)} target menus to process.")

    # 1. Fetch the side menu structure from the template
    cur.execute("SELECT id, name, limitCount, required FROM side_menus WHERE menu_id = %s", (TEMPLATE_MENU_ID,))
    template_side_menus = cur.fetchall()

    if not template_side_menus:
        print(f"No side menus found for template menu ID: {TEMPLATE_MENU_ID}")
        exit()

    print(f"Found {len(template_side_menus)} side menus in template.")

    # 2. Loop through each target menu item
    for target_menu_id in TARGET_MENU_IDS:
        print(f"\nApplying side menus to menu_id: {target_menu_id}")

        # 3. Loop through the template's side menus
        for original_side_menu_id, name, limit_count, required in template_side_menus:

            # 4. Create a NEW side_menu for the target menu and get its new ID
            new_side_menu_id = str(uuid.uuid4())
            cur.execute(
                """
                INSERT INTO side_menus (id, name, limitCount, required, menu_id)
                VALUES (%s, %s, %s, %s, %s)
                """,
                (new_side_menu_id, name, limit_count, required, target_menu_id)
            )
            print(f"  - Created new side_menu '{name}' with new ID: {new_side_menu_id}")

            # 5. Fetch all options from the original side menu
            cur.execute("SELECT name, amount, price, type FROM side_menu_options WHERE side_menu_id = %s", (original_side_menu_id,))
            template_options = cur.fetchall()

            # 6. Create new side_menu_options linked to the NEW side_menu_id
            for opt_name, opt_amount, opt_price, opt_type in template_options:
                cur.execute(
                    """
                    INSERT INTO side_menu_options (id, name, amount, price, type, side_menu_id)
                    VALUES (%s, %s, %s, %s, %s, %s)
                    """,
                    (str(uuid.uuid4()), opt_name, opt_amount, opt_price, opt_type, new_side_menu_id)
                )

            print(f"    - Added {len(template_options)} options to side menu '{name}'")

    # Commit all the changes to the database
    conn.commit()
    print(f"\n✅ Successfully applied all side menus to {len(TARGET_MENU_IDS)} target menus!")

except Exception as e:
    print(f"❌ An error occurred: {e}")
    if conn:
        conn.rollback()  # Roll back changes if anything fails
        print("Changes rolled back.")
finally:
    # Clean up connections
    if cur:
        cur.close()
    if conn:
        conn.close()
    print("Database connections closed.")
```
