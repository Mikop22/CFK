```python
import psycopg2
from datetime import date

def get_target_menu_ids(connection, restaurant_id, subsection_id=None):
    """
    Get menu IDs based on restaurant_id and optional subsection_id

    Args:
        connection: psycopg2 connection object
        restaurant_id: The restaurant ID to filter by
        subsection_id: Optional subsection ID to filter by (None means no filter)

    Returns:
        List of menu IDs
    """
    cursor = connection.cursor()

    try:
        if subsection_id is not None:
            # Query with both restaurant_id and subsection_id filters
            query = """
                SELECT menu_id
                FROM menus
                WHERE restaurant_id = %s
                AND subsection_id = %s
                AND DATE(created_at) = %s
                ORDER BY created_at DESC;
            """
            cursor.execute(query, (restaurant_id, subsection_id, date.today()))
        else:
            # Query with only restaurant_id filter (subsection_id can be anything)
            query = """
                SELECT menu_id
                FROM menus
                WHERE restaurant_id = %s
                AND DATE(created_at) = %s
                ORDER BY created_at DESC;
            """
            cursor.execute(query, (restaurant_id, date.today()))

        # Fetch all results and extract menu IDs
        results = cursor.fetchall()
        target_menu_ids = [row[0] for row in results]

        return target_menu_ids

    except psycopg2.Error as e:
        print(f"Database error: {e}")
        return []
    finally:
        cursor.close()

# Example usage:
def main():
    # Database connection parameters
    conn_params = {
        'host': 'localhost',
        'database': 'your_database',
        'user': 'your_username',
        'password': 'your_password'
    }

    try:
        # Establish connection
        conn = psycopg2.connect(**conn_params)

        # Example 1: Get menus for a specific restaurant and subsection
        restaurant_id = "your-restaurant-id-here"
        subsection_id = "your-subsection-id-here"
        target_menu_ids = get_target_menu_ids(conn, restaurant_id, subsection_id)

        # Example 2: Get all menus for a restaurant (regardless of subsection)
        # target_menu_ids = get_target_menu_ids(conn, restaurant_id)

        print("Target Menu IDs:")
        for menu_id in target_menu_ids:
            print(f"  {menu_id}")

    except psycopg2.Error as e:
        print(f"Connection error: {e}")
    finally:
        if conn:
            conn.close()

```
